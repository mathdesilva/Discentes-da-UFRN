```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# data2019 = pd.read_excel('discentes-2019.xlsx')
# data2018 = pd.read_excel('discentes-2018.xlsx')
# data2017 = pd.read_excel('discentes-2017.xlsx')
# data2016 = pd.read_excel('discentes-2016.xlsx')
# data2015 = pd.read_excel('discentes-2015.xlsx')
# data2014 = pd.read_excel('discentes-2014.xlsx')
# data2013 = pd.read_excel('discentes-2013.xlsx')
# data2012 = pd.read_excel('discentes-2012.xlsx')
# data2011 = pd.read_excel('discentes-2011.xlsx')
# data2010 = pd.read_excel('discentes-2010.xlsx')
# data2009 = pd.read_excel('discentes-2009.xlsx')
# data = pd.concat([data2019, data2018, data2017, data2016, data2015, data2014, data2013, data2012, data2011, data2010, data2009])
# data = data[ (data.nivel_ensino == "GRADUAÃ‡ÃƒO") | (data.nivel_ensino == "MESTRADO") | (data.nivel_ensino == "DOUTORADO") | (data.nivel_ensino == "GRADUAÇÃO") ]

data = pd.read_excel("data.xlsx").drop("Unnamed: 0",axis=1)

data = data.replace(to_replace ="GRADUAÃ‡ÃƒO", value ="GRADUAÇÃO")
data = data.replace(to_replace ="NÃ£o Informado", value ="Não Informado")
data = data.replace(to_replace ="IndÃ­geno", value ="Indígeno")
data = data.replace(to_replace ="CENTRO DE CIÃŠNCIAS HUMANAS, LETRAS E ARTES", value ="CENTRO DE CIÊNCIAS HUMANAS, LETRAS E ARTES")
data = data.replace(to_replace ="CENTRO DE CIÃŠNCIAS EXATAS E DA TERRA", value ="CENTRO DE CIÊNCIAS EXATAS E DA TERRA")
```

```
graduacao = data[ data.nivel_ensino == "GRADUAÇÃO" ]
posgraduacao = data[ data.nivel_ensino != "GRADUAÇÃO" ]
```

##Diferenças de genero nos cursos de graduação e pós-graduação

Analisando ambos os gráficos abaixo, podemos concluir que apesar de mínima a diferença entro os sexos, no gráfico de graduação a quantidade de sexo masculino é ligeiramente maior que a de sexo feminino, situação que se inverte no de pós-graduação.

```
fig = plt.figure(figsize = (13, 13))

ax1 = fig.add_subplot(121)
ax1 = graduacao['sexo'].value_counts().plot.pie( autopct = '%.1f%%',colors = ['orange', 'blue'])
plt.ylabel('')
plt.title("GENEROS EM CURSOS DE GRADUAÇÃO")

ax2 = fig.add_subplot(122)
ax2 = posgraduacao['sexo'].value_counts().plot.pie(autopct = '%.1f%%', colors = ['blue', 'orange'])
plt.ylabel('')
plt.title("GENEROS EM CURSOS DE PÓS-GRADUAÇÃO")

plt.show()
```

##Diferenças de gênero  no decorrer dos anos na UFRN

Fazendo a analise da diferença de sexo entre os gráficos ao passar dos anos, notamos:

* Em ambos os gráficos, tanto na pós-graduação quanto na graduação a diferença proporcional entre os sexos não é muito relevante.

* Apesar de pequena, a diferença ao longo dos anos de cada gráfico é um pouco mais estável quando se trata dos cursos de pós-graduação.


```
fig = plt.figure(figsize = (16, 13))

ax1 = fig.add_subplot(211)
ax1 = sns.countplot(x = 'ano_ingresso', hue = 'sexo', data = graduacao)
plt.title("Diferença de gênero na graduação da UFRN entre 2009 e 2019")

ax2 = fig.add_subplot(212)
ax2 = sns.countplot(x = 'ano_ingresso', hue = 'sexo', data = posgraduacao)
plt.title("Diferença de gênero na pós-graduação da UFRN entre 2009 e 2019")

plt.show()
```

## Diferença entre gêneros nos cursos de humanas e exatas

Olhando ambos os gráficos podemos concluir:

* A diferença de sexo entre os cursos de humanas é baixa, com o sexo feminino estando em maior quantidade.

* Já na área de exatas a diferença é gritante com o sexo masculino ocupando a maior parte, quase 70%.

```
humanas = data[ data.nome_unidade_gestora == "CENTRO DE CIÊNCIAS HUMANAS, LETRAS E ARTES" ]
exatas = data[ data.nome_unidade_gestora == "CENTRO DE CIÊNCIAS EXATAS E DA TERRA" ]

fig = plt.figure(figsize = (13, 13))

ax1 = fig.add_subplot(121)
ax1 = humanas['sexo'].value_counts().plot.pie( autopct = '%.1f%%',colors = ['orange', 'blue'])
plt.ylabel('')
plt.title("GENEROS EM CURSOS DE HUMANAS")

ax2 = fig.add_subplot(122)
ax2 = exatas['sexo'].value_counts().plot.pie(autopct = '%.1f%%', colors = ['blue', 'orange'])
plt.ylabel('')
plt.title("GENEROS EM CURSOS DE EXATAS")

plt.show()
```

## Diferença entre gêneros nos cursos de exatas no decorrer dos anos

Analisando esses gráficos notamos que, com o decorrer dos anos a diferença gritante que existe de sexos nos cursos de exatas não diminui de maneira relevante e não mostra um sinal de mudança para o futuro.

```
fig = plt.figure(figsize = (13, 6))
sns.countplot(x = 'ano_ingresso', hue = 'sexo', data = exatas)
plt.title("DIFERENÇA ENTRE GÊNEROS NOS CURSOS DE EXATAS AO LONGO DOS ANOS")
plt.show()
```

## Etnias nos cursos de graduação e pós-graduação

Realizando uma análise geral dos gráficos de raça entre graduação e pós-graduação pontuamos que:

* A um predomínio de brancos e pardos sobre as outras raças em ambos os gráficos, porém, nos cursos de pós-graduação a quantidade de brancos é maior em relação aos pardos, situação essa diferente nos cursos de graduação onde ocorre um equilíbrio bem maior entre as duas raças.

* A proporção de negros em ambos os gráficos são baixas, porém, são muito parecidas. 

```
fig = plt.figure(figsize = (13, 13))

ax1 = fig.add_subplot(121)
ax1 = graduacao['raca'].value_counts().plot.pie( autopct = '%.1f%%', colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple', 'tab:brown', 'tab:pink', 'tab:gray', 'tab:olive', 'tab:cyan'] )
plt.ylabel('')
plt.title("ETNIAS NOS CURSOS DE GRADUAÇÃO")

ax2 = fig.add_subplot(122)
ax2 = posgraduacao['raca'].value_counts().plot.pie( autopct = '%.1f%%', colors = ['tab:orange', 'tab:blue', 'tab:green', 'tab:red', 'tab:purple', 'tab:brown', 'tab:pink', 'tab:gray', 'tab:olive', 'tab:cyan'])
plt.ylabel('')
plt.title("ETNIAS NOS CURSOS DE PÓS-GRADUAÇÃO")

plt.show()
```

##Diferença entre etnias no decorrer dos anos na UFRN

Ao analisarmos etnia/raça nos dois gráficos e a maneira com que ela se comporta ao longo dos anos notamos que:

* Primeiramente enquanto a quantidade de pardos em cursos de graduação tem sido maior que brancos, desde 2014, nos cursos de pós-graduação essa taxa se inverte.
* Outra informação que fica clara é como taxa de não informados diminui em ambos os gráficos ao longo dos anos.
* Mais uma coisa que pode-se notar é que a variação das taxas ao longo dos anos nos cursos de pós-graduação
tende a ser mais estável que as de graduação onde ocorrem variações bem mais destoantes ao longo dos anos

```
fig = plt.figure(figsize = (16, 13))

ax1 = fig.add_subplot(211)
ax1 = sns.countplot(x = 'ano_ingresso', hue = 'raca', data = graduacao)
plt.title("Diferença entre etnias na graduação da UFRN entre 2009 e 2019")

ax2 = fig.add_subplot(212)
ax2 = sns.countplot(x = 'ano_ingresso', hue = 'raca', data = posgraduacao)
plt.title("Diferença entre etnias na pós-graduação da UFRN entre 2009 e 2019")

plt.show()
```
